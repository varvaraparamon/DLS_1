```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
#  
# 1) on-hot-encoding
dummy_features = pd.get_dummies(data[cat_cols], drop_first=True) # ,   drop_first
dummy_features.head()
```

```python
#   ,    "No phone service"  "No internet service"
#       HasPhoneService_Yes, HasInternetService_No
columns_to_drop = [col for col in dummy_features.columns if col.endswith(('No phone service', 'No internet service'))]
columns_to_drop
```

```python
dummy_features.drop(columns=columns_to_drop, inplace=True)
dummy_features
```

```python
# 
#   ,  totalspent    clientperiod    monthlyspending   
X = pd.concat([num_features.drop(columns='TotalSpent'), dummy_features], axis=1)
```

```python
y = data['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
#  
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#  LogisticRegressionCV
Cs = [100, 10, 1, 0.1, 0.01, 0.001]

model = LogisticRegressionCV(Cs=Cs, cv=5, scoring='roc_auc', refit=True, random_state=42, max_iter=1000)

model.fit(X_train, y_train)
```

```python
y_pred_proba = model.predict_proba(X_test)[:, 1]

#  ROC-AUC
roc_auc = roc_auc_score(y_test, y_pred_proba)
print(f'ROC-AUC   : {roc_auc:.4f}')
print(f'  C: {model.C_}')
```

```python
# ROC-AUC   : 0.8455
#   C: [1.]
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop(columns=['Churn']), data['Churn'], test_size=0.2, shuffle=True)
```

```python
!pip install catboost
import catboost # : https://catboost.ai/docs
```

```python
boosting_model_without_gs = catboost.CatBoostClassifier(n_estimators=200,
                                             cat_features=cat_cols)

boosting_model_without_gs.fit(X_train, y_train)

y_train_predicted = boosting_model_without_gs.predict_proba(X_train)[:, 1]
y_test_predicted = boosting_model_without_gs.predict_proba(X_test)[:, 1]
```

```python
train_auc = roc_auc_score(y_train, y_train_predicted)
test_auc = roc_auc_score(y_test, y_test_predicted)

plt.figure(figsize=(10,7))
plt.plot(*roc_curve(y_train, y_train_predicted)[:2], label='train AUC={:.4f}'.format(train_auc))
plt.plot(*roc_curve(y_test, y_test_predicted)[:2], label='test AUC={:.4f}'.format(test_auc))
legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
legend_box.set_facecolor("white")
legend_box.set_edgecolor("black")
plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
plt.show()
```

```python
#    grid search
boosting_model = catboost.CatBoostClassifier(n_estimators=200, logging_level='Silent',
                                             cat_features=cat_cols,
                                             eval_metric='AUC')
boosting_model.grid_search(
    {
        'l2_leaf_reg': np.linspace(0, 1, 20),
        'learning_rate': [0.01, 0.05, 0.1],
        'depth': [4, 6, 8]
    },
    X_train,
    y_train,
    cv=5,
    plot=True,
    refit=True,
    verbose=True
)
```

```python
y_train_predicted = boosting_model.predict_proba(X_train)[:, 1]
y_test_predicted = boosting_model.predict_proba(X_test)[:, 1]
```

```python
train_auc = roc_auc_score(y_train, y_train_predicted)
test_auc = roc_auc_score(y_test, y_test_predicted)

plt.figure(figsize=(10,7))
plt.plot(*roc_curve(y_train, y_train_predicted)[:2], label='train AUC={:.4f}'.format(train_auc))
plt.plot(*roc_curve(y_test, y_test_predicted)[:2], label='test AUC={:.4f}'.format(test_auc))
legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
legend_box.set_facecolor("white")
legend_box.set_edgecolor("black")
plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
plt.show()
```
